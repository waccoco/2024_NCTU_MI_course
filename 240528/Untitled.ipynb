```python
from matminer.datasets import load_dataset
from matminer.featurizers.composition import ElementProperty
from matminer.featurizers.conversions import StrToComposition
import pandas as pd

# 加載 steel_strength 數據集
steel_strength_df = df

# 將公式字符串轉換為 Composition 對象
converter = StrToComposition(target_col_id='composition')
steel_strength_df = converter.featurize_dataframe(steel_strength_df, 'formula')

# 定義要提取的元素屬性
element_properties = ['Number', 'MendeleevNumber', 'AtomicWeight', 'MeltingT', 'Column', 'Row', 'CovalentRadius', 'Electronegativity']
stats = ['mean']

# ep_feat = ElementProperty(preset_name='magpie', features=element_properties)
ep_feat = ElementProperty(data_source='magpie', features=element_properties, stats=stats)


# 提取元素屬性作為特徵
steel_strength_df = ep_feat.featurize_dataframe(steel_strength_df, col_id='composition')

# 查看特徵化後的數據集
print(steel_strength_df.head())

# 保存特徵化後的數據集到本地文件
steel_strength_df.to_csv('steel_strength_featurized.csv', index=False)
```
