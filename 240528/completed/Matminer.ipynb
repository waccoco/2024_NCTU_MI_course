```python
import pandas as pd
from pymatgen.core import Composition
from matminer.featurizers.composition import ElementProperty

# 假設這是你的資料集
df = pd.DataFrame({
    'composition': ['Fe2O3', 'Al2O3', 'SiO2', 'MgO', 'TiO2']
})

# 將材料組成從字符串轉換為 Composition 對象
df['composition'] = df['composition'].apply(Composition)

# 初始化 Featurizer
ep_featurizer = ElementProperty.from_preset('magpie')

# 應用 Featurizer 進行特徵轉換
df = ep_featurizer.featurize_dataframe(df, col_id='composition', ignore_errors=True)

# 顯示轉換後的數據
print("特徵轉換後的數據：")
display(df.head())
print(set(df))
```
