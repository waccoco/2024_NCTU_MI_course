```python
from matminer.datasets import load_dataset
from matminer.featurizers.conversions import StrToComposition
from matminer.featurizers.composition import ElementProperty
from matminer.featurizers.composition import ElementFraction
import pandas as pd

# 下載 steel_strength 資料集
df = load_dataset('steel_strength')

# 將 formula 轉換為 Composition 對象
df = StrToComposition().featurize_dataframe(df, 'formula')

# 使用 ElementProperty 來提取元素屬性特徵
element_properties = ['Number', 'MendeleevNumber', 'AtomicWeight', 'MeltingT', 'Column', 'Row', 'CovalentRadius', 'Electronegativity']
stats = ['mean']
# stats = ['mean', 'std_dev', 'minimum', 'maximum', 'range']
ep_featurizer = ElementProperty(data_source='magpie', features=element_properties, stats=stats)
df = ep_featurizer.featurize_dataframe(df, 'composition', ignore_errors=True)
ef_featurizer = ElementFraction()
df = ef_featurizer.featurize_dataframe(df, 'composition', ignore_errors=True)

# 刪除原始的 formula 和 composition 列
df = df.drop(columns=['formula', 'composition'])

# 保存處理後的資料集到 CSV 文件
df.to_csv('steel_strength_features.csv', index=False)

# 顯示處理後的資料集的前幾行
display(df.head())
print(set(df))
```
